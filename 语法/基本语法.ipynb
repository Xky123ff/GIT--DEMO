```python
x = 1    # int
y = 2.8  # float
z = 1j   # complex

# 从 int 转换为 float:
a = float(x)

# 从浮点数转换为整数:
b = int(y)

# 从 int 转换为 complex:
c = complex(x)

print(a)
print(b)
print(c)

print(type(a))
print(type(b))
print(type(c))
```

```python
#字符串是数组
a = "Hello, World!"
print(a[1])
```

```python
#获取从位置 2 到位置 5（不包括）的字符:
b = "Hello, World!"
print(b[2:5])
```

```python
#获取从位置 5 到位置 1 的字符，从字符串末尾开始计数:

b = "Hello, World!"
print(b[-5:-2])
```

```python
#len() 函数返回字符串的长度:

a = "Hello, World!"
print(len(a))
```

```python
#strip() 方法删除开头和结尾的空白字符:

a = "               Hello, World! "
print(a.strip()) # returns "Hello, World!"
```

```python
#lower() 返回小写的字符串:

a = "Hello, World!"
print(a.lower())
```

```python
a = "Hello, Horld!"
print(a.replace("H", "J"))
```
